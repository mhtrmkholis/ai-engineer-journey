```python
import pandas as pd

df = pd.read_csv("../data/spotify_songs.csv")
print(df.columns)

# Soal 1 — Filter & Count (easy) 
result1 = len(df[(df["energy"] > 0.9) & (df["danceability"] > 0.8)])
# print(f"Soal 1: {result1}")

# Soal 2 — Sorting & Top-N (easy-medium)
result2 = df[["track_name", "track_artist", "duration_ms"]].sort_values(by="duration_ms", ascending=False).head(5)
# print(f"Soal 2: {result2}")

# Soal 3 — GroupBy Single Agg (medium)
result3 = df.groupby("playlist_genre")["valence"].mean().sort_values(ascending=False).round(3)
# print(f"Soal 3: {result3}")

# Soal 4 — GroupBy Multi-Agg (medium)
result4 = df.groupby("playlist_genre").agg(
  n=("track_id", "count"),
  avg_popularity=("track_popularity", "mean"),
  median_popularity=("track_popularity", "median")
).round(2)
# print(f"Soal 4: {result4}")

# Soal 5 — Pivot via Unstack (harder)
result5 = df.groupby(["playlist_genre", "playlist_subgenre"]).size().unstack(fill_value=0)
# print(f"Soal 5: {result5}")

# Soal 6 — Combined Challenge (hardest)
result6 = df.groupby("track_artist").agg(
  n_tracks=("track_id", "count"),
  avg_pop=("track_popularity", "mean")
).query("n_tracks >= 10").sort_values(by="avg_pop", ascending=False).head(10).round(2)
# print(f"Soal 6: {result6}")

# Stretch Goal (kalau 6 di atas terlalu mudah)
result7 = df.groupby(["playlist_genre"])[["danceability", "energy", "valence", "acousticness", "track_popularity"]].mean().round(3)
# print(f"Soal 7: {result7}")
print(result7.shape)
print(result7.head(20))
```
