```python
import numpy as np
import csv

# Load CSV ke NumPy array (cara primitif — biasanya Pandas yang handle, tapi kita belajar)
data = []
with open("../data/spotify_songs.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        try:
            data.append(
                [
                    float(row["danceability"]),
                    float(row["energy"]),
                    float(row["loudness"]),
                    float(row["tempo"]),
                    int(row["track_popularity"]),
                ]
            )
        except (ValueError, KeyError):
            continue

features = np.array(data)
print(f"Shape: {features.shape}")
print(f"Columns: danceability, energy, loudness, tempo, track_popularity")

# Stats dasar - vectorized
print(f"\nMean per column: {features.mean(axis=0)}")
print(f"Std per column: {features.std(axis=0)}")
print(f"Min per column: {features.min(axis=0)}")
print(f"Max per column: {features.max(axis=0)}")

# Filter: lagu populer (popularity > 80)
popular = features[features[:, 4] > 80]
print(f"\nLagu populer: {popular.shape[0]} dari {features.shape[0]}") 
print(f"Avg energy lagu populer: {popular[:, 1].mean():.3f}")
print(f"Avg energy lagu lainnya: {features[features[:, 4] <= 80, 1].mean():.3f}")

# Korelasi danceability vs energy
corr = np.corrcoef(features[:, 0], features[:, 1])
print(f"\nCorrelation danceability vc energy: {corr[0, 1]:.3f}")

```
